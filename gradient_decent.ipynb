Gradient_Decent Method and its comparison

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
import math

#Grad_decent Function
def grad_decent(x,y):
    #initial slop and intercept
    n=len(x)
    m=1.5
    b= 1.0
    learning_rate_m= 0.0001
    learning_rate_b=0.0003
    #iteration= 200
    
    while(True):
        #print('iteration number :',(i+1))
        y_new= [((m*x[i])+b) for i in range(len(x))]
        #print(y_new)
        cost= sum([((y_new[i]-y[i])**2) for i in range(len(y))])/n
        # print(cost)
        w1= (y_new-y)*(x)
        #print(w1)
        w2= (y_new-y)
        #print(w2)
        d_z_m= (2*sum(w1))/n
        #print(d_z_m)
        d_z_b= (2*sum(w2))/n 
        #print(d_z_b)
        m= m-(learning_rate_m*d_z_m)
        b= b-(learning_rate_b*d_z_b)
        y_new= [((m*x[i])+b) for i in range(len(x))]
        cost2=sum([((y_new[i]-y[i])**2) for i in range(len(y))])/n
        if(math.isclose(cost,cost2,abs_tol=1e-20)):
            break
    
    y_new= [((m*x[i])+b) for i in range(len(x))] 
    cost= sum([((y_new[i]-y[i])**2) for i in range(len(y))])/n
    #print(cost) 
    return y_new
        
    
        

df=pd.read_csv('test_scores.csv') 
a=df['math']
b=df['cs']
x=np.array(a)
y=np.array(b)
y_new= grad_decent(x,y)
reg=linear_model.LinearRegression()
reg.fit(df[['math']],df['cs'])
m=reg.coef_
c=reg.intercept_
y_linear_reg=[((m*x[i])+c) for i in range(len(x))]
print(y)
print(y_linear_reg)
print(y_new)
plt.plot(x,y)
plt.plot(x,y_linear_reg)
plt.plot(x,y_new)
plt.show()
```
